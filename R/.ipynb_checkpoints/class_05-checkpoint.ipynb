# Introdução à Probabilidade

* Leitura Recomendada: Triola, 2004 capítulo 03. 
* Leitura Recomendada: Bussab e Moretim, 2017 capítulo 05. 

Probabilidade é o elemento essencial ao processo de inferência estatítica. 

Como um exemplo simples, suponha que você ganhasse o prêmio máximo em uma loteria estadual cinco vezes consecutivas. Haveria investigações e acusações de que você estava de alguma forma trapaceando. As pessoas sabem que, embora existe a possibilidade de alguém ganhar cinco vezes consecutivas com sorte honesta, essa chance é tão incrivelmente baixa que eles rejeitariam a sorte como uma explicação razoável nação. É exatamente assim que os estatísticos pensam: eles rejeitam explicações baseadas em probabilidades muito baixas. Estatísticos usam a *regra de evento raro*. [Tradução livre de (Triola, 2004)]


# Regra do Evento Raro. 

Se, sob uma dada premissa (como uma loteria ser justa), a probabilidade de um
determinado evento observado (como cinco vitórias consecutivas na loteria) é extremamente
pequeno, concluímos que a suposição provavelmente não está correta. [Tradução livre de (Triola, 2004)]

Nota: apesar da probabilidade ser pequena, há uma possibilidade de ocorrência (Ainda: 
"eventos raros ocorrem com frequência!!!")

* Um evento é qualquer coleção de resultados ou resultados de um procedimento.
* Um evento simples é um resultado ou um evento que não pode ser mais dividido em componentes mais simples.
* O espaço amostral para um procedimento consiste em todos os eventos simples possíveis. Naquela ou seja, o espaço de amostra consiste em todos os resultados que não podem ser divididos
em nenhum mais longe.

##  Exemplo
Lançar um dado e obter 5. 

5 é um evento simples e o espaço amostral é o conjunto de todas as possibilidades de valores para o lançamento de um dado $\Omega=\{1,2,3,4,5,6\}$

##  Exemplo
Lançar um dado duas vezes e obter 7.

7 não é um evento simples e o espaço amostral é o conjunto de todas as possibilidades de valores para o lançamento de um dado $\Omega=\{1-1,1-2,\cdots, 6-6\}$

# Definição de probabilidades

Notação:

* P: refere-se a probabilidade;
* A, B, ....: refere-se aos eventos 
* P(A), P(B),...: refere-se às probabilidades dos eventos A, B, ....


## Aproximação de Probabilidade por meio de frequências relativas ($fr_i$)

Lembre-se da definição de frequência relativa $fr_i=\dfrac{f_i}{\sum_{i=1}^Nf_i}$

Realize (ou observe) um procedimento várias vezes e conte os número de vezes que o evento A realmente ocorre. Com base nesses resultados reais, $P (A)$ é estimado da seguinte forma:

\begin{equation}
P(A)\approx \dfrac{n.vezes A ocorreu}{N observações}
\end{equation} 

N é um número grande de realizações/observações. 

* Nota: o termo grande é subjetivo e não é definido. 

## Abordagem Clássica 

Abordagem clássica da probabilidade (requer igualmente Resultados prováveis)

Suponha que um determinado procedimento tenha $n$ diferentes eventos simples e que cada um esses eventos simples têm a mesma chance de ocorrer. Se o evento A puder ocorrer em $s$ destas $n$ maneiras, então:

\begin{equation}
P(A)=\dfrac{\textrm{número de maneiras pelas quais A pode ocorrer}}{\textrm{número de diferentes eventos simples}}=\dfrac{s}{n}
\end{equation}

ou 

\begin{equation}
P(A)=\dfrac{\textrm{#}A}{\textrm{#}\Omega}=\dfrac{s}{n}
\end{equation}

* Observe que as duas abordagens são adequadas para diferentes situações. 

###  O lançamento da moeda e calculo de obter cara
$$P(cara)=\dfrac{1}{2}$$
A={cara}; $\Omega=${cara, coroa}.
Note que:

* A Regra 2 exige resultados igualmente prováveis. 
  * Se os resultados não forem igualmente prováveis, devemos usara estimativa de frequência relativa
* A Regra 1 fornece uma estimativa da probabilidade em vez de um valor exato.

## Lei dos grandes números

Como o número total de observações aumenta, as estimativas correspondentes tendem a se aproximar da probabilidade real. 

*Se um procedimento é repetido repetidamente, a probabilidade relativa de frequência (Regra 1) de um evento tende a se aproximar da probabilidade real.*

* A lei dos grandes números nos diz que a frequência relativa estimada a partir de A regra 1 tende a melhorar com mais observações. 
* Esta lei reflete uma noção simples apoiado pelo senso comum: uma estimativa de probabilidade baseada em apenas alguns ensaios pode ser catastrófica
* Um número de observações grande também não garante, sem sombra de dúvida, que as probabilidades serão adequadas ( A lei dos grandes números combinada com a Regra do evento Raro torna o valor aceitável e concluímos que o evento mais provável é o que está ocorrendo.)

$\lim_{N\to\infty}fr=P(A), \hspace{1.0cm}\textrm{ou}\hspace{1.0cm}f_r\to P(A)$

* Quanto mais observações, melhor a aproximação. 


## Exemplos

* Uma moeda é lançada 700 vezes e 500 caras são obtidas. Qual a  probabilidade aproximada para obter cara em um lançamento ?
A={Cara}
$P(A)\approx \dfrac{500}{700}$

* Se a moeda anterior é honesta, então podemos aplicar probabilidade clássica para calcular a probabilidade de obter cara em um lançamento:

$P(\{Cara\})=\dfrac{1}{2}$

* Uma pesquisa selecionou uma amostra aleatória de 855 pesssoas. O espaço amostral consiste em dois eventos simples: selecionar uma pessoa e verificar se voou em um avião comercial ou não. Porque o espaço da amostra consiste em eventos que não são igualmente prováveis, não podemos usar a abordagem clássica (Regra 2). Podemos usar a abordagem de frequência relativa (Regra 1) sabendo que entre 855 adultos selecionados aleatoriamente, 710 indicaram que eles voaram em aviões comerciais.  [Tradução livre de (Triola, 2004) ]

A={A pessoa voou em avião comercial}

$P(A)\approx \dfrac{710}{855}$

*  Você planeja apostar no número 13 na próxima rodada de uma roda de roleta. Qual é a probabilidade de você perder?

  * assumindo eventos igualmente prováveis. 
  
  A={não sair 13 na casela}
  
  $$P(A)=\dfrac{37}{38}$$
  
* Encontre a probabilidade de que quando um casal que têm 3 filhos, eles terão exatamente 2 meninos. Suponha que meninos e meninas são igualmente prováveis e que o sexo de qualquer criança não é influenciado pelo gênero de qualquer outra criança.
  
  A={2 filhos em 3 nascimentos}
  
  $$P(A)=\dfrac{{3}\choose{2}}{2^3}=3/8=0.375$$


## Resultados interessantes sobre Probabilidade

* $0 \leqslant P(A)\leqslant 1$
* Se $P(A)=0$: A é evento impossível
* Se $P(A)=1$: A é evento certo (irá ocorrer)
* Evento complementar
  *  *O complemento do evento A, indicado por A, consiste em todos os resultados em que evento A não ocorre.*

* Exemplo: Em um grupo típico, existem 205 recém-nascidos, 105 dos quais são meninos. Se um
o bebê é selecionado aleatoriamente do grupo, qual é a probabilidade de que o bebê não seja um menino?

A={O bebê é um menino}; $A^c=\overline{A}=\{\textrm{O bebê não é menino}\}=\{\textrm{O bebê é menina}\}$

\begin{equation}
P(A^c)=\dfrac{205-105}{205}=1-\dfrac{105}{205}=1-P(A)
\end{equation}

  * $P(A^c)=1-P(A)$
  * $P(A^c)+P(A)=1$ ou $P(\Omega)=1$ pois $A\cup A^c=\Omega$

## Princípio Fundamental da Contagem

* Se a tarefa 1 pode ser realizada de m maneiras e a tarefa 2 pode ser realizada de n maneiras, então o número de maneiras de realizar as tarefas 1 e 2 é mn. 
* O princípio anterior pode ser generalizado para um número k de tarefas. 
  * se $n_1, n_2,\cdots, n_k$ são, respectivamente, o número de maneiras de realizar k tarefas, então o número total de possibilidades para realização das k tarefas é:
  
  $N=n_1\cdot n_2\cdots \cdot n_k$
  
*  Suponha que num lote com 20 peças existam cinco defeituosas. Escolhemos quatro peças do lote ao acaso, ou seja, uma amostra de quatro elementos, de modo que a ordem dos elementos seja irrelevante. Determine a probabilidade de obter 2 peças defeituosas em uma amostra aleatória de 04 peças. 


* O Princípio Fundamental da Contagem dá origem às conhecidas fórmulas de combinações, arranjos e permutações:
    * Combinações: $C_{n,k}=\dfrac{n!}{k!(n-k)!}={{n}\choose{k}}={{n}\choose{n-k}}$
    * Arranjos: $A_{n,k}=k!C_{n,k}$
    * Permutações (simples e com elementos repetidos)
      * $P_{n,k}=n\cdot(n-1)\cdots(n-(k+1))=\dfrac{n!}{(n-k))!}$
      * $P_{n,k}=n^k$
