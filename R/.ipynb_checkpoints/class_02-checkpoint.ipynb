# Ordenação de dados ou Rol

* Representação dos dados ordenados ou Rol: $x_1 , x_2, . . . ,x_n$

As notas dos 50 alunos ordenadas:
$3.67, 9.58, 11.49, 14.77, ... 92.37, 94.38, 98.75$

O Mínimo e o Máximo: $x_1= 3.67$ e $x_{50}= 98.7$

Amplitude do Rol: $A = x_{50} − x_1= 95.1$

*** 

* O Rol é importante para identificar dados discrepantes ou Outlier. No entanto, um método gráfico pode ter mais apelo visual.

* Quais outras informações podem ser obtidas pela análise do Rol?

* Uma visualização melhor dos dados pode ser obtida por meio da tabela de Frequências.

# Tabela de frequências

Definição: Uma distribuição de frequência (ou tabela de frequência)
lista os valores dados individualmente ou por intervalos
(ou classes) juntamente com sua frequência (contagem)

Para construir uma tabela de frequência:

* Limite inferior de classe ($l_i$): são so menores números de cada classe.
* Limite superior de classe ($L_i$): são os maiores números de cada classe.
* Fronteiras de classe: são os números usados para separar as classes, podem pertencer ou não a classe.
* Ponto médio das classes ($x_i$): são os pontos médios de cada intervalo que determina a classe.
* Amplitude da classe ($A_c$): é a diferença entre os limites inferior e superior de cada classe ou a diferença entre as fronteiras de cada classe.
* Número de classes ($N_c$): regra de Sturges 

$$N_c= 1 + 3.3\log_{10}(N)$$

## Exemplos

### Exemplo 
Considere que em um conjunto de dados ordenados temos $x_1 =155$, $x_n= 190$.

Amplitude Rol: $A = 190 − 155 = 35$. 

Note que 35 é divisível por 7 ou por 5. Podemos optar por $N_c = 5$ ou $N_c = 7$.

Amplitude das classes: $N_c= 7$, $A_c=\dfrac{A}{N_c}=\dfrac{35}{7}=5$. 
    
    

### Exemplo 
Considere os dados das 70 notas dos alunos anteriormente mencionados.

```r
set.seed(1000)
options(digits=4)
x=100*runif(70)
x=sort(x)
N=length(x)
Nc=ceiling(1+3.33*log10(N)); 
A=round(max(x)-min(x),1);
Ac=round(A/Nc,1); 
breaks = seq(min(x),max(x),Ac) 

breaks=c(breaks,max(x)+0.1)
x.cut = cut(x, breaks, right=FALSE) 
x.freq = table(x.cut)
cbind(x.freq)

cat("total",sum(x.freq))
```

Nota: Cuidado ao utilizar softwares para construir a distribuição de frequências. A distribuição pode estar incorreta! Verificações são sempre necessárias. Veja o código a seguir!

```r
x=10*runif(250)
options(digits=4)
N=length(x)
Nc=floor(1+3.3*log10(N)); 
A=max(x)-min(x);
Ac=round(A/Nc,1); 
breaks = seq(min(x),max(x), by=Ac)   
breaks=c(breaks,max(x)+0.1)
x.cut = cut(x, breaks, right=FALSE) 
x.freq = table(x.cut) 
cbind(x.freq) 
cat("total",sum(x.freq))
```

* Não trazem informação suficiente para reprodução dos dados. 
* Se existem classes com zero elementos, é indicado obter uma outra distribuição com intervalos de classe. 

## Procedimento para construção da tabela de frequências

Para construir uma tabela de frequência:

* Passo 1: Ordenação dos dados $x_1, x_2, . . . , x_n$,
* Passo 2: Calcular a amplitude do Rol: $A = x_n-x_1$
* Passo 3: Calcular o número de classes (aproximadamente) $N\approx 1 + 3.3\log_{10} (N)$ ,
* Passo 4: Calcular a amplitude das classes $A_c \approx A/N_c$
* Passo 5: Calcular os limites inferior e superior de cada classe
* Passo 6: Contagem das observações dentro de cada classe

# Outras Frequências 

* Frequência Relativa: $Fr=\dfrac{f_i}{\sum_{i=1^k}f_i}$
* Frequência Relativa Percentual: $Fr_p=100*Fr$
* Frequência Acumulada: $Fa_i=\sum_{k=1}^{i}f_k$
* Frequência Relativa Acumulada: $Fra_i=\sum_{k=1}^{i}Fr_k$

## Exemplos

### Exemplo 
150 dados gerados de uma população normal com $\mu=0$ e $\sigma^2=1.0$

```r
x=rnorm(150)
options(digits=4)
N=length(x)
Nc=floor(1+3.3*log10(N)); #Nc
A=max(x)-min(x);
Ac=A/Nc; #Ac
breaks = seq(min(x)-0.25,max(x)+0.5, by=Ac)    #half-integer sequence 
x.cut = cut(x, breaks, right=FALSE) 
f_i = table(x.cut) 

#-----outras frequencias
options(digits=2)
Fr_i=f_i/sum(f_i)#frequencia relativa
Frp_i=100*Fr_i#frequencia relativa percentual
Fa_i=cumsum(f_i) #frequencia acumulada
Fra_i=cumsum(Fr_i)##frequencia relativa acumulada
Frap_i=100*Fra_i#frequencia relativa acumulada
cbind(f_i,Fr_i,Fa_i,Fra_i, Frp_i,Frap_i) 
sum(f_i)
```

### Exemplo

## Construção Manual do Histograma

* Frequências ($f_i$)
  * Construir tabela de frequências ($f_i$) 
  * No intervalo $[l_{inf_i},l_{sup_i})$ o valor é $f_i$ (Função constante no intervalo)
      * um retângulo de base $[l_i,L_i)$ e altura $f_i$ 
  * Repetir o processo para todos os intervalos
* Frequências relativas ($Fr_i$)
  * Construir tabela de frequências ($Fr_i$) 
  * No intervalo $[l_{inf_i},l_{sup_i})$ o valor é $Fr_i$ (Função constante no intervalo)
      * um retângulo de base $[l_i,L_i)$ e altura $Fr_i$ 
  * Repetir o processo para todos os intervalos
* Frequência relativa percentual ($Frp_i$)
  * Idem processos anteriores

# Discussão complementar

Histograma, polígono de frequências e a função de densidade.

* Histograma e pol. frequências: $N_1$ valores

```r
options(digits=4)
set.seed(100)
x=c(rnorm(20,175.0,4.0))

h = hist(x,breaks=5,probability=TRUE,col = "Steelblue3", right = F,main="",xlab="alturas",ylab="prob.",plot=TRUE,xlim=c(160,190),ylim=c(0.,0.250))
#h
#atributes from h: breaks,counts,density,mids,xname,equidist,attr(,"class")
#cat("limites:",h$breaks)
#------pontos médios
delta=h$mids[2] - h$mids[1];
mp = c(min(h$mids)-delta ,h$mids, max(h$mids) + delta)
#freq = c(0, h$counts, 0)
freqrel = c(0, h$density, 0)
#--------------adicionando polígono ao histograma
#h = hist(s, xlim = c(0, max(s) + 10), col = "Steelblue3", right = F)
lines(mp, freqrel, type = "b", pch = 20, col = "red", lwd = 3)

#-------------curva verdadeira
x1 <- seq(160, 190, length=100)
hx <- dnorm(x1,175.,sd=4.0)
hxest <- dnorm(x1,mean(x),sd=sd(x))

par(new=TRUE)
plot(x1, hx, type="l", lty=2, xlab="",
  ylab=" ", main="",xlim=c(160,190),col="red",ylim=c(0.,0.250))
par(new=TRUE)
plot(x1, hxest, type="l", lty=2, xlab="",
  ylab=" ", main="",xlim=c(160,190),ylim=c(0.,0.250))
```

Dados numéricos sem intervalo de classe (Gráfico de Barras): Amostra de tamanho $N=100$ retiradas de $\{0,1,\cdots,10\}$ com reposição. 

```r
options(digits=4)
set.seed(100)
#x=c(rnorm(500,175.0,4.0))
x=sample(0:10,50,replace=TRUE)
x.freq=table(x)
cbind(x.freq)
barplot(x.freq,xlim=c(0,15),col = "Steelblue3",border="red")#h#Tirar o # e executar
#cat("limites:",h$breaks)#idem linha anterior
```

## Exemplo 
Dados categóricos(Gráfico de Barras): Dados **mtcars** e tipo de transmissão marcha (automático ou manual). 

```r
options(digits=4)
#set.seed(100)
#x=c(rnorm(500,175.0,4.0))
x=mtcars$am
x.freq=table(x)
names=factor(x)
cbind(x.freq)
barplot(x.freq,col = "Steelblue3",border="red")#h#Tirar o # e executar
#cat("limites:",h$breaks)#idem linha anterior
```

# Diagrama de dispersão


## Exemplo
Diagrama dispersão: $hp\times wt$

```r
options(digits=4)
y=mtcars$wt
x=mtcars$hp
plot(hp~wt, # Specify the variables to graph
        data = mtcars, #Specify the data set
        pch = 19, #Changes the plotting symbol
        main = "mtcars data", # Create a title for the plot
        xlab = "wt", #Create the x-axis label
        ylab = "hp") #Create the y-axis label
cat("A correlação entre as variáveis é",cor(x,y))
```

* Parece haver uma relação linear entre as variáveis *wt* e *hp* ? 

## Exemplo
Diagrama dispersão: $hp\times qsec$

```r
options(digits=4)
y=mtcars$qsec
x=mtcars$hp
plot(hp~qsec, # Specify the variables to graph
        data = mtcars, #Specify the data set
        pch = 19, #Changes the plotting symbol
        main = "mtcars data", # Create a title for the plot
        xlab = "wt", #Create the x-axis label
        ylab = "hp") #Create the y-axis label
cat("A correlação entre as variáveis é",cor(x,y))
```

* Parece haver uma relação linear entre as variáveis *wt* e *hp* ?

# Exemplo: 
Apresentaremos a seguir algumas representações e análises com o R utilizando uma base de dados aberta. Utilizamos a base de dados sobre Licitações realizadas no Brasil em Janeiro de 2019. 

```r
jan_2019<-read.delim("dados/licitacao.csv",header = FALSE, sep = ";",encoding = "latin1")
jan_2019<-jan_2019[-c(1),]
colnames(jan_2019)<-c("Número Licitação","Número Processo","Objeto","Modalidade Compra","Situação Licitação","Código Órgão Superior", "Nome Órgão Superior","Código Órgão", "Nome Órgão", "Código UG","Nome UG", "Município", "Data Resultado Compra", "Data Abertura"," Valor Licitação ")
head(jan_2019)
```

## Separando as variáveis

Podemos manipular os dados de acordo com as nossas necessidades. No nosso exemplo, como a base de dados é relativamente grande, pretendemos separar o aquivo único que temos, em arquivos separáveis, um para cada órgão que realiza a Licitação. Segue o código utilizado.

```r
lista_vazia<-c() #cria uma variável que será utilizada para armazenamento.

lista_orgaos<-unique(jan_2019$`Nome Órgão`) #coleta uma unica vez os Órgãos cadastrados
# 
head(lista_orgaos)
# 
##jan_2019#essa linha está dando problema pois esses dados estão sendo mostrados na tela/impressos
# 
lenghData<-length(jan_2019)

for (i in 1:lenghData) {

  lista_vazia<-subset(jan_2019,jan_2019$`Nome Órgão`==lista_orgaos[i])
  #write.table(lista_vazia,lista_orgaos[i])
  }#função para ler e comparar os arquivos
lista_vazia
```

## Algumas análises
 
* Por modalidade de compra

```r
library(ggplot2)
ggplot(jan_2019,aes(x=(jan_2019$`Modalidade Compra`)))+
  geom_bar()+
  theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1))+
  labs(x="Modalidade", y="Quantidade", title = "MODALIDADES DE LICITAÇÕES")
```

##  Analisando por Situação

```r
ggplot(jan_2019,aes(x=(jan_2019$`Situação Licitação`)))+
  geom_bar()+
  theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1))+
  labs(x="Situação", y="Quantidade", title = "SITUAÇÃO DA LICITAÇÃO")
```

##  Analisando por Comando

###  Comando da Marinha

```r
marinha<-read.csv("dados/Comando da Marinha",sep = "")
marinha<-marinha[,-c(1,2,3,5,13,14)]
marinha[order(marinha$Valor.Licitação),]

ggplot(marinha,aes(x=reorder(Nome.UG, Valor.Licitação),y=Valor.Licitação))+
geom_bar(stat="identity")+
  theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1))+
  labs(x="OM", y="Valor da Licitação", title = "LICITAÇÕES")
  #geom_text(aes(label=Valor.Licitação), vjust=-1) (inserir legenda)
```

### Instituto Brasileiro de Museus

```r
inst_bras_museus<-read.csv("dados/Instituto Brasileiro de Museus", sep = " ")
inst_bras_museus
inst_bras_museus<-inst_bras_museus[,-c(1,2,3,5,13,14)]
inst_bras_museus[order(inst_bras_museus$Valor.Licitação),]

ggplot(inst_bras_museus,aes(x=reorder(Nome.UG, Valor.Licitação),y=Valor.Licitação))+
  geom_bar(stat="identity")+
  theme(axis.text.x = element_text(angle = 45, vjust = 1,hjust = 1))+
  labs(x="OM", y="Valor da Licitação", title = "LICITAÇÕES")
```

## Formatando o gráfico de barras
Percebe-se nos exemplos acima que os gráficos construídos necessitam de alguns ajustes: aumentar a fonte dos rótulos dos eixos, centralizar o título, mudar as cores, etc. Utilizaremos um dos gráficos já construídos para ilustrar essas alterações.

```r
ggplot(jan_2019,aes(x=(jan_2019$`Modalidade Compra`)))+
  geom_bar(color="black", fill="yellow")+
  theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1, size = 18), axis.text.y = element_text(size=18),axis.title.x = element_text(size=20),axis.title.y = element_text(size=20),plot.title = element_text(hjust = 0.5, size = 20))+
  labs(x="Modalidade", y="Quantidade", title = "MODALIDADES DE LICITAÇÕES")
```
