## As regras da adição  e outros itens interessantes

* $0\leqslant P(A)\leqslant 1$
* $P(\emptyset)=0$ (evento impossível), $P(\Omega)=1$ (evento certo)
* $P(A\cup B)=P(A)+P(B)$ se $A\cap B=\emptyset$
  * $A\cap B=\emptyset$ ou dizemos que A e B são disjuntos se eles não podem ocorrer ao mesmo tempo.
* $A \subset B$, então $P(A)\leqslant P(B)$
* $P(A\cup B)=P(A)+P(B)-P(A\cap B)$ se $A\cap B\neq \emptyset$

### Exemplos
####  
Consideremos um experimento aleatório e os eventos A e B associados, tais que P(A) = 1/2, P(B) = 1/3 e P(A \cap B) = 1/4. Então temos:
* $P(A^c) = 1 – P(A) = 1 – 1/2 = 1/2$
* $P(B^c) = 1 – P(B) = 1 – 1/3 = 2/3$.
* $P(A\cup B) = P(A) + P(B) – P(A \cap B) = 1/2 + 1/3 – 1/4 = 7/12$.
* $P(A^c \cap B^c ) = P[(A\cup B)^c ] = 1 – P(A\cup B) = 1 – 7/12 = 5/12$.
* $P(A^c\cup B^c ) = P[(A\cap B)^c ] = 1 – P(A\cap B) = 1 – 1/4 = 3/4$.
* Calculemos $P(A^c\cap B)$, isto é, a probabilidade de que ocorra B e não ocorra A.
Podemos escrever
$B = (A\cap B) \cup A^c \cap B),$

Como $$ P(\Omega)=\sum_{i=1}^{n}P(\{\omega_i\})=1\Rightarrow P(\{\omega_i\})=1/n$$.


Indique  $M$ o evento que ocorre quando um aluno do conjunto  é escolhido ao acaso e  ele for um estudante de Matemática Pura. A, E, C, H e F têm significados análogos. Assim $P(A) = 30/200$, ao passo que $P(H) = 115/200$ conforme ilustrado na tabela. 

* Calcule as outras probabilidades. 
* Determine $P(A\cup H)$, ou seja, a probabilidade de ser do Curso de Matemática Aplicada e ser homem. 
* Determine todas as outras possibilidades de uniões dois a dois. 

#### Mostre a fórmula para $P(A\cup B\cup B)$

### Exercícios

#### Identifique se os eventos abaixo são mutuamente exclusivos. 
* a)
    * Selecionando aleatoriamente um cirurgião cardíaco.
    * Selecionando aleatoriamente uma médica.
* b) 
  * Selecionando aleatoriamente uma estudante universitária
  * Selecionando aleatoriamente um estudante universitário que dirige uma motocicleta
* c. 
  * Selecionando aleatoriamente alguém tratado com o medicamento para redução de colesterol 
  * Selecionando aleatoriamente alguém em um grupo de controle que não recebeu nenhum medicamento

####  Uma fábrica produz determinado artigo. Da linha de produção são reti rados três artigos, e cada um é classificado como bom (B) ou defeituoso (D). Um espaço amostral do experimento é $\Omega= {BBB, BBD, BDB, DBB, DDB, DBD, BDD, DDD}$. Se A designar o evento que consiste em obter dois artigos defeituosos, então $A = \{DDB, DBD, BDD\}$. 
  * Determine a probabilidade de obter 0,1,2,3 defeitos.
  * Determine a probabilidade de obter pelo menos dois defeitos.
  * Determine a probabilidade de obter no máximo dois defeitos.

#### Duas moedas são lançadas. Liste os eventos seguintes e calcule as respectivas probabilidade:
  * pelo menos uma cara;
  * duas caras;
  * o complementar do evento anterior.

####  Uma moeda e um dado são lançados. Dê um espaço amostral do experimento e depois
represente-o como produto cartesiano dos dois espaços amostrais, correspondente aos
experimentos considerados individualmente.
  * Obtenha a probabilidade de obter uma cara com um 6.
  * Obtenha a probabilidade de obter uma coroa com um número menor que 4.
  
#### Dentre seis números positivos e oito negativos, dois números são escolhidos ao acaso (sem reposição) e multiplicados. Qual a probabilidade de que o produto seja positivo?

#### Considere o lançamento de dois dados. Considere os eventos: A = soma dos números obtidos igual a 9, e B = número no primeiro dado maior ou igual a 4. Enumere os elementos de A e B. Obtenha $A \cup B$, $A\cap B$ e $A^c$ e calcule as respectivas probabilidades. 

#### Use os dados na tabela a seguir, que resume os resultados do afundando do Titanic. 

|    |Homens|Mulheres|Meninos|Meninas|
|:---:|:----:|:---:|----:|:---:|
Sobreviveu|332|318| 29|27
Faleceu|1360|104| 35|18|

  *  Se um dos passageiros do Titanic for selecionado aleatoriamente, encontre a probabilidade de que o resultado seja mulher ou criança.
  * Se um dos passageiros do Titanic for selecionado aleatoriamente, encontre a
probabilidade de que o resultado seja um homem ou alguém que sobreviveu ao naufrágio.
  * Se um dos passageiros do Titanic for selecionado aleatoriamente, encontre a
probabilidade de que o resultado seja um menino ou alguém que sobreviveu ao naufrágio.
  * Se um dos passageiros do Titanic for selecionado aleatoriamente, encontre a
probabilidade de que o resultado seja uma mulher ou alguém que não sobreviveu ao naufrágio.


## A regra da multiplicação

Considere o exemplo: Encontre a probabilidade de um casal ter pelo menos 1 menina entre 3 crianças. Suponha que meninos e meninas sejam igualmente provável e que o gênero de uma criança seja independente do gênero de qualquer irmão irmãs ou irmãs.

Pelo Princípio Multiplicativo, temos: 

$$P(A)=\dfrac{1}{8}=\dfrac{1}{2} \cdot \dfrac{1}{2} \cdot \dfrac{1}{2} $$

* Quando $A$ e $B$ são independentes $$P(A\cap B)=P(A)\cdot P(B)$$
Isso significa que: Somente quando $A$ e $B$ são independentes, o fato de $A$ ocorrer não afeta $B$ e vice-versa. 
* Nota: Eventos disjuntos e independentes são coisas distintas:
  $A\cap B=\emptyset \Rightarrow P(A\cap B)=0$, enquanto que $0\leqslant P(A\cap B)=P(A)\cdot P(B)\leqslant 1$
  
<!--- Inserir anotação de aula aqui
Como outro exemplo de independência, considere uma prova com apenas duas questões. A primeira questão consiste em anotar uma das alternativas 

(Ver anotações de sala)

--->

A seguir, consideramos o segundo ponto principal desta seção, baseado no princípio de que a probabilidade de um evento é frequentemente afetada pelo conhecimento de circunstâncias. Por exemplo, se você selecionar aleatoriamente alguém da população geral a probabilidade de conseguir um homem é $0,5$ devido à suposição de equiprobabilidade de eventos. No entanto,  se você souber que a  pessoa muda frequentemente de canal de TV com um controle remoto, a probabilidade é 0,999 (Uma suposição exagerada talvez!). [Tradução livre de (Triola, 2004, pg. 151)]

## Probabilidade  Condicional  e independência

A *probabilidade condicional* de um evento ocorre quando a probabilidade é afetada pelo conhecimento de outras circunstâncias. A probabilidade condicional do evento $B$ ocorrer, dado que o evento $A$ tem já ocorreu, pode ser encontrada usando a regra de multiplicação 

$$P(A \hspace{0.1cm}e\hspace{0.1cm}  B) =P (A)\cdot P (B|A)$$ 
que conduz a:
$$P (B|A)=\dfrac{P(A \hspace{0.1cm}e\hspace{0.1cm} B)}{P(A)}=\dfrac{P(A \cap B)}{P(A)}$$ 

### Exemplos


#### (Triola, 2004, pg. 152) Ensaios clínicos do teste de gravidez: A tabela a seguir apresenta dados de um teste de gravidez realizado com um grupo de mulheres.  Supondo que 1 pessoa é selecionada aleatoriamente dentre as 99 pessoas incluídas no estudo,  encontre a probabilidade de selecionar um sujeito que esteja grávida ou teve um resultado positivo.

Resultados |positivo "+"|negativo "-"|
 |:-----:|:----:|:----:|
grávida (g)| 80| 5|
não grávida(ng)| 3|11|

A tabela acima é muito informativa, pois torna possível o calculo de várias probabilidades tais como $P(grávida), P(não grávida), P("+")$  e $P("-")$. Além disso, permite avaliar as probabilidades condicionais tais como:

* $P("+"|grávida)=\dfrac{P("+"\hspace{0.1cm}e\hspace{0.1cm} grávida)}{P(grávida)}=\dfrac{80}{85}$

* $P("+"|não grávida)=\dfrac{P("+"\hspace{0.1cm}e\hspace{0.1cm} grávida)}{P(não grávida)}=\dfrac{3}{14}$

* $P("-"|grávida)=\dfrac{P("+"\hspace{0.1cm}e\hspace{0.1cm} grávida)}{P(grávida)}$

* $P("-"|não grávida)=\dfrac{P("-"\hspace{0.1cm}e\hspace{0.1cm} grávida)}{P(não grávida)}$

* $P(grávida|"+")$
* $P(grávida|"-")$
* $P(não grávida|"+")$
* $P(não grávida|"-")$


Nota: é possível observar que a tabela de frequências torna possível calcular todas as probabilidades e probabilidades condicionais por meio da simples observação dos valores e aplicação da fórmulas. 


#### Reconsiderando a tabela: 


```{r fig01_1, echo=FALSE, fig.cap="Número de alunos matrículados em cada curso. Fonte: Bussab e Moretim, Estatística Básica, 6 ed. pg. 107", out.width = '90%'}
knitr::include_graphics("Fig01.png")
```

Queremos $P(C_3|B)=\dfrac{C_3\cap B}{P(B)}$ (Teorema de Bayes)

#### Exemplo

Para selecionar seus funcionários, uma empresa oferece aos candidatos um curso de treinamento durante uma semana. No final do curso, eles são submetidos a uma prova e 25% são classificados como bons (B), 50% como médios (M) e os restantes 25% como fracos (F). Para facilitar a seleção, a empresa pretende substituir o treinamento por um teste contendo questões referentes a conhecimentos gerais e específicos. Para isso, gostaria de conhecer qual a probabilidade de um indivíduo aprovado no teste ser considerado fraco, caso fizesse o curso. No final do curso, obtiveram-se as seguintes probabilidades condicionais:

$P(A|B)=0.8$, $P(A|M)=0.5$, $P(A|F)=0.2$

Queremos 

$P(F|A)=\dfrac{P(F\cap A)}{P(A)}=\dfrac{P(F\cap A)}{P(B)\cdot P(A|B)+P(M)\cdot P(A|M)+P(F)\cdot P(A|F)}$ 

Logo, 

$P(F|A)=\dfrac{(0,20)(0,25)}{(0,80)(0,25) + (0,50)(0,50) + (0,20)(0,25)}= 0,10.$
